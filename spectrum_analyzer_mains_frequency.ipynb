# Analyzing spectrum of mains frequency with Arduino UNO 

## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import plotting
%matplotlib inline
```

## Parameters

```python
fs = 2427 # sampling frequency 
bit = 10 # arduino resolution
resolution = 5 / 2**10 # arduino range: 0...5 Volt
```

## Read Data from A/D Converter

```python
data = np.genfromtxt('analog_data.txt') * resolution
```

## Plot Time Domain

```python
plotting.plot_time(data, Fs);
plt.ylim([0,5])
plt.xlim([0,0.5])
```

## Plot Frequency Domain

### linear and around 50 Hz

```python
plotting.plot_freq(data[:300000], Fs);
plt.ylim([0, 2])
plt.xlim([49.8, 50.2]);
```

### linear with harmonics

```python
plotting.plot_freq(data[:300000], Fs);
plt.ylim([0, 0.1])
plt.xlim([1, 400]);
```

### dB with harmonics

```python
plotting.plot_freq(data[:300000], Fs, scale='db');
plt.xlim([1, Fs/2]);
```
